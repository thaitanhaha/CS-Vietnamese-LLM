```python
import torch
from unsloth import FastLanguageModel

MAX_SEQ_LENGTH = 2048
DTYPE = None
LOAD_IN_4BIT = True

model_general, tokenizer_general = FastLanguageModel.from_pretrained(
    "/kaggle/input/cs-vietnamese/pytorch/default/2/chatbot_model",
    max_seq_length=MAX_SEQ_LENGTH,
    load_in_4bit=LOAD_IN_4BIT,
    dtype=DTYPE,
)

model_general_1, tokenizer_general_1 = FastLanguageModel.from_pretrained(
    "/kaggle/input/cs-vietnamese/pytorch/default/2/chatbot_model_2",
    max_seq_length=MAX_SEQ_LENGTH,
    load_in_4bit=LOAD_IN_4BIT,
    dtype=DTYPE,
)

model_cs, tokenizer_cs = FastLanguageModel.from_pretrained(
    "/kaggle/input/cs-vietnamese/pytorch/default/2/cs_model",
    max_seq_length=MAX_SEQ_LENGTH,
    load_in_4bit=LOAD_IN_4BIT,
    dtype=DTYPE,
)
```

```python
from transformers import GenerationConfig

generation_config_general = GenerationConfig(
    max_new_tokens=2048,
    temperature=1.0,
    do_sample=False,
    num_return_sequences=1,
    pad_token_id=tokenizer_general.eos_token_id,
    eos_token_id=tokenizer_general.eos_token_id,
    repetition_penalty=1.3
)

generation_config_general_1 = GenerationConfig(
    max_new_tokens=2048,
    temperature=1.0,
    do_sample=False,
    num_return_sequences=1,
    pad_token_id=tokenizer_general_1.eos_token_id,
    eos_token_id=tokenizer_general_1.eos_token_id,
    repetition_penalty=1.3
)

generation_config_cs = GenerationConfig(
    max_new_tokens=2048,
    temperature=1.0,
    do_sample=False,
    num_return_sequences=1,
    pad_token_id=tokenizer_cs.eos_token_id,
    eos_token_id=tokenizer_cs.eos_token_id,
    repetition_penalty=1.3
)

def gen(prompt, model, tokenizer, generation_config, is_cs=False):
    if is_cs:
        processed_prompt = [
            {"role": "system", "content": ("Bạn là một trợ lý AI chuyên trả lời các câu hỏi về lĩnh vực lập trình, khoa học máy tính,... Hãy trả lời bằng tiếng Việt.")},
            {"role": "user", "content": (prompt)},
        ]
    else:
        processed_prompt = [
            {"role": "system", "content": ("Bạn là một trợ lý AI thân thiện, hãy trả lời bằng tiếng Việt.")},
            {"role": "user", "content": (prompt)},
        ]
        
    chat_text = tokenizer.apply_chat_template(
        processed_prompt,
        add_generation_prompt=True,
        tokenize=False
    )
    
    inputs = tokenizer(
        chat_text,
        return_tensors="pt"
    ).to("cuda:0")
    
    with torch.no_grad():
        outputs = model.generate(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"],
            generation_config=generation_config,
        )
    output_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    
    if "assistant\n" in output_text:
        answer = output_text.split("assistant\n")[-1].strip()
    else:
        answer = output_text.strip()
    
    return answer
```

```python
import pandas as pd 

test_df = pd.read_csv("/kaggle/input/cs-qas/cs_qa.csv")
test_df
```

```python
prompts = test_df['Question']
general_responses = []
general_1_responses = []
cs_responses = []

for prompt in prompts:
    print(f"Prompt: {prompt}")
    
    print('-' * 50)
    general_response = gen(prompt, model_general, tokenizer_general, generation_config_general, is_cs=False)
    general_responses.append(general_response)
    print(f"Response General: {general_response}")
    
    print('-' * 50)
    general_response_1 = gen(prompt, model_general_1, tokenizer_general_1, generation_config_general_1, is_cs=True)
    general_1_responses.append(general_response_1)
    print(f"Response General: {general_response_1}")
    
    print('-' * 50)
    cs_response = gen(prompt, model_cs, tokenizer_cs, generation_config_cs, is_cs=True)
    cs_responses.append(cs_response)
    print(f"Response CS: {cs_response}")
    
    print("\n" * 3)
    print('-*' * 50)
```

```python
gpt_responses = test_df['Answer']

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from bert_score import score

scores = np.zeros((len(prompts), 3))

for i in range(len(prompts)):
    _, _, F1_general = score([gpt_responses[i]], [general_responses[i]], lang="vi")
    _, _, F1_general_1 = score([gpt_responses[i]], [general_1_responses[i]], lang="vi")
    _, _, F1_cs = score([gpt_responses[i]], [cs_responses[i]], lang="vi")

    scores[i, 0] = F1_general.item()
    scores[i, 1] = F1_general_1.item()
    scores[i, 2] = F1_cs.item()

plt.figure(figsize=(12, 8))
sns.heatmap(scores, annot=True, cmap="YlGnBu", xticklabels=["model_general", "model_general_1", "model_cs"], yticklabels=prompts)
plt.title("BERTScore F1 Heatmap of Models by Prompt")
plt.xlabel("Models")
plt.ylabel("Prompts")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
