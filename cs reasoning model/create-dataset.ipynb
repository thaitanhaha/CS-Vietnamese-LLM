```python
import pandas as pd
from datasets import load_dataset
from huggingface_hub import login

login(token="MY_TOKEN")
dataset = load_dataset("diwank/python-code-execution-output")
```

```python
df_train = pd.DataFrame(dataset['train'])
df_test = pd.DataFrame(dataset['test'])
df_train.head()
```

```python
import python_minifier

code_str = """
def greet(name):
    print(name)
    print(12)
"""

print(python_minifier.minify(code_str, remove_literal_statements=True, rename_globals=True, remove_asserts=True, remove_debug=True))
```

```python
df_train['new_code'] = df_train['code'].apply(
    lambda x: python_minifier.minify(
        x, remove_literal_statements=True, rename_globals=True, remove_asserts=True, remove_debug=True
    )
)
df_test['new_code'] = df_test['code'].apply(
    lambda x: python_minifier.minify(
        x, remove_literal_statements=True, rename_globals=True, remove_asserts=True, remove_debug=True
    )
)

df_train
```

```python
df_train.to_csv('code_with_output_train.csv', index=False)
df_test.to_csv('code_with_output_test.csv', index=False)
```
